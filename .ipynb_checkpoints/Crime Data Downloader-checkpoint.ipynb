```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
import io
```

```python
data_frame_list = []
```

```python
for i in range(0, 3):
    page = requests.get(f'https://estadisticas.pr/en/estadisticas-mas-recientes?type=delitos_tipo_i&page={i}', 
                        verify = False)
    
    page = bs(page.content, 'html.parser')
    
    spans = page.findAll('span', class_ = 'file')
    
    for span in spans:
        try:
            date = span.find('a').text
            link = span.find('a')['href']

            data = requests.get(link, verify = False)

            output = open('data.xls', 'wb')
            output.write(data.content)
            output.close()

            df = pd.read_excel('data.xls', sheet_name='MUNICIPIOS')
            df = df.melt(id_vars = ['Distrito'])
            df.query('Distrito != "TOTAL"', inplace=True)
            df['Fecha'] = date
        except:
            continue

        data_frame_list.append(df)
```

```python
df = pd.concat(data_frame_list)
```

```python
df.head()
```

### Loading months with inconsistencies

```python
marzo_2016 = pd.read_excel('data/marzo_2016/TIPO I POR MUNICIPIOS 2016.xls')
marzo_2016.drop('Unnamed: 10', axis = 1, inplace=True)
marzo_2016.dropna(inplace=True)
marzo_2016 = marzo_2016.melt(id_vars = ['Distrito'])
marzo_2016['Fecha'] = 'Marzo 2016'
marzo_2016.head()
```

```python
nov_2015 = pd.read_excel('data/noviembre_2015/TIPO I POR MUNICIPIOS 2015.xls')
nov_2015.drop('Unnamed: 10', axis = 1, inplace=True)
nov_2015.dropna(inplace=True)
nov_2015 = nov_2015.melt(id_vars = ['Distrito'])
nov_2015['Fecha'] = 'Noviembre 2015'
nov_2015.head()
```

```python
abril_2012 = pd.read_excel(
    'data/abril_2012/13 REGIONES ABRIL 2012.xls',
    sheet_name='MUNICIPIOS',
    header=1)

abril_2012 = abril_2012.melt(id_vars = ['Distrito'])
abril_2012['Fecha'] = 'Abril 2012'
abril_2012.head()
```

### Joining all dataframes together

```python
final_df = pd.concat([df, marzo_2016, nov_2015, abril_2012])
final_df.query('Distrito not in ["Distrito", "TOTAL"]', inplace=True)
final_df.head()
```

```python
final_df.to_csv('type_I_crimes.csv', index=False)
```
